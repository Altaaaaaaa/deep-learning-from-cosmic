```python
import numpy as np

def tanh(x):
    p_exp_x = np.exp(x)
    m_exp_x = np.exp(-x)

    y = (p_exp_x - m_exp_x)/(p_exp_x + m_exp_x)

    return y
```

```python
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)

# 캔버스 설정
fig = plt.figure(figsize=(10, 7)) # 캔버스 생성
fig.set_facecolor('white') # 캔버스 색상 설정

plt.plot(x, y)
plt.title("Hyperbolic Tangent", fontsize=30)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20, rotation=0)

plt.yticks([-1.0, 0.0, 1.0]) # 특정 축에서 특정 값만 나오게
plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True) # y축에 있는 모든 숫자에 회색 점근선을 그음

plt.show()
```

```python
# 하이퍼볼릭 탄젠트의 미분
def diff_tanh(x):
    return 4 / (np.exp(2*x) + 2 + np.exp(-2*x))
```

```python
import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y1 = diff_tanh(x)

fig = plt.figure(figsize = (10, 5))

plt.plot(x, y1, label = "diff_tanh")

plt.title("tanh", fontsize=30)
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize=20, rotation=0)

plt.ylim(-0.5, 2)
plt.xlim(-7, 7)

plt.legend(loc = "upper right")

plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True)
ax.xaxis.grid(True)

plt.show()
```
