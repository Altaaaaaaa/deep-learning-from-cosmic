**Sigmoid Function**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative 

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.001)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = sigmoid(x)
ax1.plot(x, y)

y = derivative(sigmoid, x)
ax2.plot(x, y)
plt.show()
```

**Tanh Function**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.arange(-10, 10, 0.001)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = tanh(x)
ax1.plot(x, y)

y = derivative(tanh, x)
ax2.plot(x, y)
plt.show()
```

**ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def ReLU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(0)
    return np.array(res)

x = np.arange(-10, 10, 0.001)
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = ReLU(x)
ax1.plot(x, y)

y = derivative(ReLU, x)
ax2.plot(x, y)
plt.show()
```

**Leaky ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def Leaky_ReLU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(0.01 * value)
    return np.array(res)

x = np.arange(-20, 10, 0.001)
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = Leaky_ReLU(x)
ax1.plot(x, y)

y = derivative(Leaky_ReLU, x)
ax2.plot(x, y)
plt.show()
```

**ELU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def ELU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(np.exp(value) - 1)
    return np.array(res)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

x = np.arange(-20, 10, 0.001)
y = ELU(x)
ax1.plot(x, y)

y = derivative(ELU, x)
ax2.plot(x, y)
plt.show()
```
