**Sigmoid Function**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative 

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.001)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = sigmoid(x)
ax1.plot(x, y)

y = derivative(sigmoid, x)
ax2.plot(x, y)
plt.show()
```

**Tanh Function**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

x = np.arange(-10, 10, 0.001)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = tanh(x)
ax1.plot(x, y)

y = derivative(tanh, x)
ax2.plot(x, y)
plt.show()
```

**ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def ReLU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(0)
    return np.array(res)

x = np.arange(-10, 10, 0.001)
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = ReLU(x)
ax1.plot(x, y)

y = derivative(ReLU, x)
ax2.plot(x, y)
plt.show()
```

**Leaky ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def Leaky_ReLU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(0.01 * value)
    return np.array(res)

x = np.arange(-20, 10, 0.001)
fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = Leaky_ReLU(x)
ax1.plot(x, y)

y = derivative(Leaky_ReLU, x)
ax2.plot(x, y)
plt.show()
```

**ELU**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def ELU(x):
    res = []
    for value in x:
        if value > 0:
            res.append(value)
        else:
            res.append(np.exp(value) - 1)
    return np.array(res)

fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

x = np.arange(-20, 10, 0.001)
y = ELU(x)
ax1.plot(x, y)

y = derivative(ELU, x)
ax2.plot(x, y)
plt.show()
```

**활성화 함수에서의 문제들**
1. Vanishing Gradient Problem
  - 음수 값 -> 0: 최종 레이어에 영향 적어짐
  - x의 절댓값이 커질수록 미분값이 소실될 가능성 커짐
2. 중심이 0이 아니다
  - 학습이 느려질 수 있다.
3. Dying ReLU Problem
  - 음수 값 -> 함수 값 = 0
  - 0보다 작은 값 뉴런 죽음

**이유 탐구**
은닉층에서는 학습을 위해 중간이 0이고 기울기 소실을 줄여야 한다<br>
-> ReLU, Leaky ReLU, ELU<br>

분류를 위해서는 각각의 확률을 구해주는 함수를 사용해야 한다<br>
-> Sigmoid, Softmax<br>

회귀를 위해서는 각각의 예측"값"을 구해주는 함수를 사용해야 한다<br>
-> Identity<br>

**Thus, 맞는 활성화 함수는 그 함수의 특성에 따라 사용한다**<br>
Ex, 소리의 파형을 분석히여 어떤 사람인지 맞추는 프로그램<br>
-> 은닉: ELU, 출력: Softmax<br>
Ex, 어떤 사람의 소리로 노래를 부르게 하는 프로그램<br>
-> 은닉: ReLU, 출력: Identity<br>
